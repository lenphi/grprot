```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Calculate current date
current_date = datetime.now()

# Calculate age for each person
def calculate_age(born):
    return (current_date - born).days // 365

uber_df['Geburtsdatum'] = pd.to_datetime(uber_df['Geburtsdatum'])
uber_df['Alter'] = uber_df['Geburtsdatum'].apply(lambda x: calculate_age(x))

result = uber_df.groupby(['Name']).agg({'Länge': 'sum', 'Alter': 'mean'})
result = result.sort_values('Länge', ascending=False)
#result = result.head(10)

fig, ax = plt.subplots()
ax.scatter(result['Alter'], result['Länge'], c='red')

# Add labels to the plot
ax.set_xlabel('Alter')
ax.set_ylabel('Länge')
ax.set_title('Correlation between Age and Sum of Lengths')

plt.show()
```

```python
# group the data by name and sort by the total length of each person
grouped = uber_df.groupby('Name').agg({'Länge': 'sum'}).reset_index()
grouped = grouped.sort_values('Länge', ascending=False)

# define the number of persons in each group
group_size = 16

# calculate the total length of all persons
total_length = grouped['Länge'].sum()

# create a dataframe to store the results
results = []

# loop through the groups
for i in range(0, len(grouped), group_size):
    group = grouped.iloc[i:i+group_size]
    group_length = group['Länge'].sum()
    percentage = group_length / total_length * 100
    results.append((group, group_length, percentage))

# print the results
for i, (group, group_length, percentage) in enumerate(results):
    print(f"Group {i+1}: {group_length} ({percentage:.2f}% of total)")
    print(group)
    print("\n")
```

```python
import matplotlib.pyplot as plt

# group the dataframe by party and calculate the sum of lengths for each party
grouped_df = uber_df.groupby("Partei")["Länge"].sum().reset_index()

# sort the dataframe by the sum of lengths
grouped_df = grouped_df.sort_values("Länge", ascending=False)

# set the labels and sizes for the pie chart
labels = grouped_df["Partei"]
sizes = grouped_df["Länge"]

# Plot the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.set_title("Added Lengths of Each Party")

# add a circle to make it a donut chart
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
# Liste der Parteien und ihrem jeweiligen Frauenanteil im Zürcher Gemeinderat zu Beginn der Legislatur 2018-2022
# Quelle: https://www.stadt-zuerich.ch
parteien_liste = [['AL', 50],
                  ['EVP', 25],
                  ['FDP', 20],
                  ['GLP', 28.6],
                  ['Grüne', 37.5],
                  ['SP', 40.5],
                  ['SVP', 18.8]]
```

```python
# Funktion, welche die Länge aller Voten von Männern und Frauen der jeweiligen Partei ausgibt
def parteien (partei):
    criteria = uber_df['Partei'] == partei
    df = uber_df[criteria]
    grouped = df.groupby('Anrede')
    result = grouped['Länge'].sum()
    result = result.reset_index()
    total = result['Länge'].sum()
    result['Percentage'] = result['Länge'] / total * 100
    return (result.sort_values(by='Länge', ascending=False))
```

```python
# Funktion, welche die RednerInnen der jeweiligen Partei und die Gesamtlänge ihrer Voten ausgibt
def parteien2 (partei):
    criteria = uber_df['Partei'] == partei
    df = uber_df[criteria]
    grouped = df.groupby('Name')
    result = grouped['Länge'].sum()
    return (result.sort_values(ascending=False))
```

```python
# Schlaufe, welche die beiden vorherigen Funktionen nutzt und über die Liste der Parteien iteriert
results = []
results2 = []
for i in parteien_liste:
    partei = i[0]
    frauenanteil = i[1]
    result = parteien(partei)
    results.append(result)
    print(partei)
    print(frauenanteil)
    print(result) 
    print()
    result2 = parteien2(partei)
    results2.append(result2)
    print(result2)
    print()
```
