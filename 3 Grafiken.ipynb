# Grafiken

## 1. Libraries und Daten laden

## 1.1 libraries

```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import seaborn
```

## 1.2 dataframe laden

```python
def uber_df_load():
    with open('uber_df_backup2.pickle', 'rb') as f:
        return pickle.load(f)
```

```python
uber_df = uber_df_load()
uber_df = uber_df[uber_df['Partei'] != 'Parteilos']
uber_df = uber_df[uber_df['Partei'] != 'FPD']
```

## 2. Grafiken

### 2.1 Redeanteil Frauen im Vergleich zu ihrem Sitzanteil nach Partei

```python
# Group the data by party and gender
grouped = uber_df.groupby(['Partei', 'Anrede'])

# Calculate the sum of the length of speeches for each group
grouped_sum = grouped['Länge'].sum().reset_index()

# Create a pivot table with the sum of the length of speeches for each party and gender
pivot_table = pd.pivot_table(grouped_sum, values='Länge', index='Partei', columns='Anrede')

# Calculate the total length of speeches for each party
total_lengths = pivot_table.sum(axis=1)

# Add a column for the percentage of speeches for each gender
pivot_table['Redeanteil Frau'] = pivot_table['Frau'] / total_lengths * 100

# Define a dictionary for the proportion of women for each party
womens_proportion_dict = {'AL': 50, 'EVP': 25, 'FDP': 20, 'GLP': 28.6, 'Grüne': 37.5, 'SP': 40.5, 'SVP': 18.8}

# Create a new dataframe with the percentage of speeches made by women and the proportion of women for each party
party_df = pivot_table.copy()
party_df["Frauenanteil der Partei im Gemeinderat"] = party_df.index.map(lambda x: womens_proportion_dict.get(x, 0))

# Plot the percentage of speeches made by women and the proportion of women for each party
party_df = party_df[party_df.index != "FPD"]
party_df["Frauenanteil der Partei im Gemeinderat"] = party_df.index.map(lambda x: womens_proportion_dict[x] if x != "FPD" else 0)

# Sort the dataframe by the percentage of speeches made by women
party_df = party_df.sort_values("Redeanteil Frau", ascending=False)

parties = party_df.index

fig, ax = plt.subplots()

bar_width = 0.4
opacity = 0.8

for i, party in enumerate(parties):
    x = i
    y1 = party_df.loc[party]["Redeanteil Frau"]
    y2 = party_df.loc[party]["Frauenanteil der Partei im Gemeinderat"]
    ax.bar(x, y1, bar_width, alpha=opacity, color='b')
    ax.bar(x + bar_width, y2, bar_width, alpha=opacity, color='r')

ax.set_xlabel("Partei")
ax.set_ylabel("Prozent")
ax.set_xticks(range(len(parties)))
ax.set_xticklabels(parties)
ax.legend([ax.bar(0, 0, color='b')[0], ax.bar(0, 0, color='r')[0]], ['Redeanteil Frauen', 'Frauenanteil der Partei im Parlament'], loc='upper left')

plt.show()
```

Nur bei der FDP (deutlich) und der GLP (knapp) liegt der Redeanteil der Frauen höher als ihr Sitzanteil. Bei der SP liegt er knapp darunter, bei allen übrigen Parteien sogar deutlich.

### Redeanteil der Parteien insgesamt (in Minuten pro Stunde)

```python
# group the dataframe by party and calculate the sum of lengths for each party
grouped_df = uber_df.groupby("Partei")["Länge"].sum().reset_index()

# Remove the 'FPD' and 'Parteilos' rows from the dataframe
grouped_df = grouped_df[~grouped_df["Partei"].isin(["FPD", "Parteilos"])]

# sort the dataframe by the sum of lengths
grouped_df = grouped_df.sort_values("Länge", ascending=False)

# set the labels and sizes for the pie chart
labels = grouped_df["Partei"]
sizes = grouped_df["Länge"]

# Create a dictionary for the colors of each party
colors = {"SP": "red", "SVP": "darkgreen", "Grüne": "green", "FDP": "blue", "AL": "magenta", "GLP": "lightgreen", "EVP": "yellow"}
colors_list = [colors[x] for x in labels]

# Plot the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct=lambda p: '{:.0f} min'.format(p * 60 / 100), startangle=90, colors=colors_list, counterclock=False)
ax.set_title("Durchschnittliche Redezeit der Parteien im Gemeinderat pro Stunde", y=1.05)

# add a circle to make it a donut chart
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal') 
plt.tight_layout()
plt.show()
```

Die SVP war mit 17 von 125 Sitzen zwar nur die fünftgrösste Partei im Gemeinderat, ihre VertreterInnen haben jedoch fast den gleichen Redeanteil wie die SP, die mit 43 Sitzen stärkste Partei im Parlament.

### Die Top-10 der Vielschwätzer im Gemeinderat

```python
# Get the total length of speeches by speaker
total_length = uber_df.groupby(["Name", "Partei", "Anrede"])["Länge"].sum().reset_index()

# Sort the data by total length of speeches
total_length = total_length.sort_values(by="Länge", ascending=False).head(10)

# Create a dictionary for mapping party to color
party_colors = {"AL": "magenta", "SVP": "darkgreen", "GLP": "lightgreen", "Grüne": "green",
                "FDP": "blue", "EVP": "yellow", "SP": "red"}

# Plot the bar chart of the top 10 speakers
bars = plt.bar(total_length["Name"] + " " + total_length["Partei"], total_length["Länge"], color=[party_colors[x] for x in total_length["Partei"]])
plt.xlabel("Name und Partei")
plt.ylabel("Länge aller Reden (Anzahl Zeichen)")
plt.xticks(rotation=90)
plt.title("Die Vielredner im Zürcher Gemeinderat")
plt.tight_layout()

# Add gender symbols to the bars
for i, bar in enumerate(bars):
    gender = total_length.iloc[i]["Anrede"]
    height = bar.get_height()
    if gender == "Herr":
        plt.text(bar.get_x() + bar.get_width() / 2, height + 100, "♂", ha="center", va="bottom", fontsize=18)
    else:
        plt.text(bar.get_x() + bar.get_width() / 2, height + 100, "♀", ha="center", va="bottom", fontsize=18)

plt.show()
```

Neun Männer und eine Frau: Diese Mitglieder des Gemeinderats haben insgesamt am meisten gesprochen von 2018 bis 2022. Besonders fleissig war Walter Angst (AL): Seine protokollierten Voten umfassen ungefäht 280 A4-Seiten.
